```python
##################################################################################################
# Patch-based Classification of Breast Cancer Histology Images using CNNs
# LE48: MiniProject
# Jan Ondras (jo356), Trinity College
# 2017/2018
##################################################################################################
######################################################################################
# Split stainnormalised images into train and validation sets 
# (no patching at this stage)
# This is for RCNN
# DONE
######################################################################################

path = './../Dataset/ICIAR2018_BACH_Challenge/Photos_SN_split' # imgs after split
org_path = './../Dataset/ICIAR2018_BACH_Challenge/Photos_SN/'   # imgs before split

# import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
# import cv2
import glob
import os
import tifffile as tiff

classes = ['Normal', 'Benign', 'InSitu', 'Invasive'] # correspond to labels 0,1,2,3 in this order
imgs = {}
imgs['train'] = {}
imgs['validation'] = {}

if not os.path.exists(path + '/train/'):
    os.mkdir(path + '/train/')
if not os.path.exists(path + '/validation/'):
    os.mkdir(path + '/validation/')

for c in classes:
    if not os.path.exists(path + '/train/' + c):
        os.mkdir(path + '/train/' + c)
    if not os.path.exists(path + '/validation/' + c):
        os.mkdir(path + '/validation/' + c)
    # Train-validation split
    imgs['train'][c], imgs['validation'][c] = train_test_split(
        glob.glob(org_path + c + '/*.tif'), random_state=32, train_size=0.75)
    print "Class ",c, " train: ", len(imgs['train'][c]), ", validation: ", len(imgs['validation'][c])
    # Copy images into corresponding folders
    for set_type in ['train', 'validation']:
        print set_type
        for img_name in imgs[set_type][c]:
            img = tiff.imread(img_name)
            print path +  '/' + set_type + '/' + c + '/' + img_name.split('/')[-1][:-4] + '.tif'
            tiff.imsave(path +  '/' + set_type + '/' + c + '/' + img_name.split('/')[-1][:-4] + '.tif', 
                               img)
                
N_train_imgs_total = len(imgs['train']['Normal']) + len(imgs['train']['Benign']) + len(imgs['train']['InSitu']) + len(imgs['train']['Invasive'])
N_val_imgs_total = len(imgs['validation']['Normal']) + len(imgs['validation']['Benign']) + len(imgs['validation']['InSitu']) + len(imgs['validation']['Invasive'])

print "Training imgs: ",  N_train_imgs_total
print "Validation imgs: ",  N_val_imgs_total
```
